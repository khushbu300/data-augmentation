```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = load_img('data/dog.jpg', target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
import numpy as np
```

```python
img_array = np.array(img)
```

```python
img_array.shape
```

```python
img_array = img_array.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(img_array, batch_size = 1, save_to_dir='aug'):
    i = i+1
    if i==10:
        break
```
